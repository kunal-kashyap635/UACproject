# Import Library

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
DATA_PATH = "../data/processed/clean_children_data.csv"
df = pd.read_csv(DATA_PATH)
df
```

# Daily Trend – Total System Load

```python
fig = px.line(
    df,
    x="date",
    y="total_system_load",
    title="Daily Total System Load",
    labels={"date": "Date", "total_system_load": "Children Under Care"},
)

fig.update_traces(line=dict(width=2))
fig.update_layout(hovermode="x unified", template="plotly_white")
fig.show()
```

# Create Rolling Averages (Pressure Smoothing)

* Daily data noisy hota hai.
* Rolling average = true system pressure.

```python
df["total_load_7d_avg"] = df["total_system_load"].rolling(7).mean()
df["total_load_14d_avg"] = df["total_system_load"].rolling(14).mean()

df["net_intake_7d_avg"] = df["net_daily_intake"].rolling(7).mean()
df["net_intake_14d_avg"] = df["net_daily_intake"].rolling(14).mean()
```

```python
df
```

# System Load Trend

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["total_system_load"],
        mode="lines",
        name="Daily Total System Load",
        line=dict(color="lightgray"),
    )
)

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["total_load_7d_avg"],
        mode="lines",
        name="7-Day Rolling Avg",
        line=dict(color="blue", width=3),
    )
)

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["total_load_14d_avg"],
        mode="lines",
        name="14-Day Rolling Avg",
        line=dict(color="red", width=3),
    )
)

fig.update_layout(
    title="Total System Load with Rolling Averages",
    xaxis_title="Date",
    yaxis_title="Number of Children",
    template="plotly_white",
)

fig.show()
```

# Net Intake Pressure Trend

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["net_daily_intake"],
        mode="lines",
        name="Daily Net Intake",
        line=dict(color="lightgray"),
    )
)

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["net_intake_7d_avg"],
        mode="lines",
        name="7-Day Avg Net Intake",
        line=dict(color="green", width=3),
    )
)

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["net_intake_14d_avg"],
        mode="lines",
        name="14-Day Avg Net Intake",
        line=dict(color="orange", width=3),
    )
)

fig.add_hline(y=0, line_dash="dash", line_color="black")

fig.update_layout(
    title="Net Intake Pressure Over Time",
    xaxis_title="Date",
    yaxis_title="Net Intake (Transfers − Discharges)",
    template="plotly_white",
)

fig.show()
```

# Care Load Volatility (Stability Indicator)

```python
df["load_volatility_7d"] = df["total_system_load"].rolling(7).std()
```

```python
df
```

```python
fig = px.line(
    df,
    x="date",
    y="load_volatility_7d",
    title="7-Day Care Load Volatility",
    labels={"load_volatility_7d": "Standard Deviation"},
    template="plotly_white",
)

fig.show()
```

# Stress Window Detection

* Logic:

1. Net intake positive
2. 7-day avg net intake positive
3. Sustained pressure

```python
df["stress_window"] = (df["net_intake_7d_avg"] > 0) & (
    df["total_load_7d_avg"] > df["total_system_load"].median()
)
```

```python
df
```

# Visualize Stress Periods

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df["date"],
        y=df["total_system_load"],
        mode="lines",
        name="Total System Load",
        line=dict(color="lightblue"),
    )
)

fig.add_trace(
    go.Scatter(
        x=df[df["stress_window"]]["date"],
        y=df[df["stress_window"]]["total_system_load"],
        mode="markers",
        name="Stress Periods",
        marker=dict(color="red", size=6),
    )
)

fig.update_layout(
    title="Detected Capacity Stress Windows",
    xaxis_title="Date",
    yaxis_title="Total System Load",
    template="plotly_white",
)

fig.show()
```

# Quantify Stress Impact

```python
stress_summary = {
    "total_days": len(df),
    "stress_days": df["stress_window"].sum(),
    "stress_percentage": (df["stress_window"].sum() / len(df)) * 100,
}

pd.DataFrame.from_dict(stress_summary, orient="index", columns=["value"])
```

### Trend & Stress Analysis Insights

- Rolling averages reveal prolonged periods of elevated system load.
- Sustained positive net intake signals backlog accumulation risk.
- Detected stress windows indicate periods of heightened capacity strain.
- These insights can inform staffing, shelter expansion, and policy timing.
